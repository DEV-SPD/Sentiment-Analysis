1. Handling Null Values 
2. Removing Punctuation
3. Spelling Correction
4. Stopwords Removal
5. Lemmatization

```python
# Importing necessary libraries
import re
from textblob import TextBlob
from spacy.lang.en import STOP_WORDS
```

```python
set_of_stop_words = set(STOP_WORDS)

new_set_stop_words = set_of_stop_words - set(['not', 'no', 'never', 'very', 'extremely', 'barely','of','why','what','is','in','it','was','have','had'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['date'])
```

```python
df = df.rename(columns={'variation':'product','verified_reviews':'reviews'})
```

```python
df.head()
```

```python
# Dropping Irrelevant Column
df = df.drop(columns=['rating','product'])
df
```

```python
# 1. Removing Punctuations
def Punctuation_remover(text2):
    Punct_pattern = re.compile(r'[^\w\s]')
    return Punct_pattern.sub('',text2) 


# 2. Lemmatization 
def lemmatize(text5):
    doc = nlp(text5)
    
    new = []
    
    for tokken in doc:
      new.append(tokken.lemma_)
    
    new_text = ' '.join(new)
    return new_text


# 3. Stop Words 
def stop_word(text6):
    new1 = [word for word in text6.split() if word.lower() not in new_set_stop_words]
    return ' '.join(new1)

# 4. Spelling Correction 
def spelling_correction(text7):
    textblb = TextBlob(text7)
    return textblb.correct().string
```

```python
df['reviews'] = df['reviews'].apply(stop_word)
df['reviews'] = df['reviews'].apply(Punctuation_remover)
df['reviews'] = df['reviews'].apply(lemmatize)
```

```python
df['reviews'] = df['reviews'].apply(spelling_correction)
```
